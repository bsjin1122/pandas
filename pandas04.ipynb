### 히스토그램: 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용하는 그래프
total_bill, tip 등의 열을 변수라고 부르기도 한다. 변수를 하나만 사용해서 그린 그래프를 `일변량 그래프`라고 한다.

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1) #그래프 기본 틀의 행 크기, 두번째는 기본 틀의 열 크기를 지정
```

```python
axes1.hist(tips['total_bill'], bins=10) #bins 인잣값으로 x축의 간격 조절 
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

## 산점도 그래프 
변수 2개를 사용해서 만든 그래프, 변수 2개를 상용하기 때문에 통계 용어로 `이변량 그래프`라고 부른다. 

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 박스 그래프 : 이산형 변수와 연속형 변수를 함께 사용하는 그래프

이산형 변수 : Female, Male과 같이 명확하게 구분 되는 값<p></p>
연속형 변수 : Tip과 같이 명확하게 셀 수 없는 범위의 값

```python
boxplot = plt.figure() #박스 그래프를 그릴 수 있다.
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 그래프
## 3개 이상의 변수를 사용한 그래프 : 다변량 그래프

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else: 
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']* 10, #s : 점의 크기 
    c = tips['sex_color'], #점의 색상
    alpha = 0.5 #인잣값을 0.5로 점의 투명도를 보기 좋게 조절
)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## seaborn 라이브러리 자유자재로 활용하기 
seaborn 라이브러리를 사용하면 보다 화려한 그래프를 그릴 수 있다.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots() #기본 틀 만들기
ax = sns.distplot(tips['total_bill']) #열을 전달하면 히스토그램 생성 완료
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# 밀집도 그래프 끄기 
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False) #밀집도 그래프 인잣값 False
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 밀집도 그래프만 나타내기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False) #hist False
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill  Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
# 양탄자 그래프(rug)는 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프이다. 
```

```python
# count 그래프
ax = plt.subplots() 
ax = sns.countplot('day', data=tips) 
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency') 
```

## seaborn 으로 산점도 그래프 그리기 

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) #회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

## 산점도 그래프와 히스토그램을 한 번에 그려주는 jointplot 메서드 사용하기

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 육각 그래프 사용하기 
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") # kind 인잣값을 hex로 지정. 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)

# 육각형으로 그래프를 그리는 이유는, 육각형이 2차원 표면을 덮을 때 가장 효과적이기 때문입니다.
```

## 이때 관계 그래프는 중복된 정보가 표현된다는 단점이 있다.
이런 경우 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체하면 된다.

```python
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) 
plt.show()
```

## seaborn 라이브러리로 그래프 스타일 설정

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('whitegrid') # 그래프의 배경에 가로줄 생김
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
#모든 스타일에 for문으로 하나씩 적용한 그래프
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() 
```
